```python
import matplotlib.pyplot as plt
```

```python
na = data["NA_Sales"].mean()
eu = data["EU_Sales"].mean()
jp = data["JP_Sales"].mean()
other = data["Other_Sales"].mean()
gs = data["Global_Sales"].mean()
```

```python
print(na, eu, jp, other, gs)
```

```python
mean = {'NA':na, 'EU':eu, 'JP':jp, "OS":other, "GS":gs}
x = list(mean.keys())
y = list(mean.values())
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(x, y, color="skyblue", width=0.4, edgecolor='blue')
plt.xlabel("Sales")
plt.ylabel("Average of Sales")
plt.title("Bar plot of average sales")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(x, y, color='pink', linewidth=3, marker='o', markersize=15, linestyle='--')
plt.xlabel("Sales")
plt.ylabel("Average of Sales")
plt.title("Bar plot of average sales")
plt.show()
```

```python
genre = data["Genre"]
```

```python
genre.unique()
```

```python
genre.value_counts()
```

```python
genre_x = genre.unique()
genre_y = genre.value_counts()
```

```python
fig = plt.figure(figsize = (15, 8))
plt.plot(genre_x, genre_y, color='green')
plt.xlabel("Genre")
plt.ylabel("Amount of Genre")
plt.title("Line plot for genre")
plt.show()
```

```python
colors = ( "orange", "cyan", "yellow", "grey", "green",)

fig = plt.figure(figsize = (20, 15))
plt.pie(genre_y, labels=genre_x, autopct='%1.2f%%', colors=colors, shadow=True)
```
